## Plot compare training loss

```python
path_ = f"Saved_Results/NIID_10client_bias_noisy/fedavg_ratio_1.0_C_0.5_.json"
with open(path_,"r") as f:
    result_dict1 = json.load(f)

path_ = f"Saved_Results/NIID_10client_bias_noisy/fedalgo1_ratio_0.8_C_0.5_.json"
with open(path_,"r") as f:
    result_dict2 = json.load(f)

```

```python

round_ = 3
# plt.figure(figsize=(20,8) figsize=(24,8), sharex=True, sharey=True)
selected_client = result_dict1[f"Round {round_}"]["selectd_client"]
fig, ax = plt.subplots(nrows=1, ncols=len(selected_client), figsize=(24,8), sharex=True, sharey=True)
for id,client in enumerate(selected_client):
    ax[id].plot(range(5), result_dict1[f"Round {round_}"]["training_loss"][f"client_{client}"], label = f"FedAvg")
    ax[id].plot(range(5), result_dict2[f"Round {round_}"]["training_loss"][f"client_{client}"], label="FedAlgo1")
    ax[id].set_title(f"Client {client}")
plt.legend()
plt.plot()
```
