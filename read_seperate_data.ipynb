```python
with open("pill_dataset/medium_pilldataset/train_dataset_samples.json", "r") as f:
    data = json.load(f)['samples']
```

```python
label_dict = {}
for i in range(150):
    label_dict[i] = {'clean': [],
                     'bright': [],
                     'zoom': [],
                     'cover': []}
```

```python
for i,item in enumerate(data):
    label_dict[item[1]][item[0].split("/")[-2].lower()].append(i)
```

```python
with open("pill_dataset/medium_pilldataset/categories.json", "w") as f:
    json.dump(label_dict, f)
```

```python
len(label_dict)
```

```python
hash_dict = {}
count = 0
clean_data = []
for key in label_dict.keys():
    label = int(key)
    list_clean = label_dict[key]['clean']
    for id in list_clean:
        hash_dict[count] = id
        count +=1 
        clean_data.append([count, label])
```
